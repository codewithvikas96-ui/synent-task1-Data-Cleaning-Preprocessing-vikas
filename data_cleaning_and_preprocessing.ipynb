#### Handle Missing Values


```python
#Handle Missing Values
missing_values = df.isna().sum()
print("Total number of missing values in each columns")
print(missing_values)

print()
print("Column that have the most missing data is: ")
col_name = missing_values.sort_values(ascending=False).index[0] 
col_val = missing_values.sort_values(ascending=False).iloc[0]
print(f"{col_name} : {col_val}")
```

```python

missing_pct = (df.isna().sum() / len(df) * 100).round(2)
print(pd.DataFrame({'Missing Count': missing_values, 'Missing %': missing_pct}))
```

```python
# Filling missing age with median over mean because meadian gives middle value 
df['age'] = df['age'].fillna(df['age'].median())

# Filling missing values in embarked and embark_town with the mode (most frequent value).
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])
df['embark_town'] = df['embark_town'].fillna(df['embark_town'].mode()[0])

#Droping the deck column
print("The deck column has too many missing values.")
print(f"Among {df.shape[0]} rows deck column have {df.isna().sum().loc['deck']} missing values")
print("Droping the deck column")

df.drop(columns=['deck'], inplace=True)
```

```python
print("Handled missing values successfully")

df.isna().sum()
```

#### Remove Duplicates

```python
print(f"There are {df.duplicated().sum()} Duplicate rows in titanic dataset")
print("Droping the duplicate rows")

df.drop_duplicates(inplace=True)

print("New Shape of the dataset after the removal of duplicate rows")
print(f"{df.shape[0]} rows x {df.shape[1]} columns")
```

#### Convert Data Types

```python
df.info()
```

```python
#Converting survived column from integer to a proper categorical type.
print("Data type of survived column:", df['survived'].dtype)
df['survived'] = df['survived'].astype('category')
print("After conversion dtype:", df['survived'].dtype)

# Converting pclass and sex columns to category dtype.
df['pclass'] = df['pclass'].astype('category')
df['sex'] = df['sex'].astype('category')

#checking and converting age,fare to float64
for col in ['age', 'fare']:
    if df[col].dtype != 'float64':
        df[col] = df[col].astype('float64')
print("age and fare dtypes confirmed:", df['age'].dtype, df['fare'].dtype)
```

```python
df.info()
```

#### Rename Columns & Final Check

```python
df.columns.to_list()
```

```python
# Rename sibsp → siblings_spouses and parch → parents_children.

df.rename(columns= {
    'sibsp' : 'siblings_spouses', 
    'parch' : 'parents_children'
 }, inplace=True)

df.columns
```

```python
# Final Check 
print("Final missing values check:")
print(df.isna().sum())
print("Final duplicate check:", df.duplicated().sum())
```

```python
# Saving the cleaned dataset as titanic_cleaned.csv for EDA 

df.to_csv('titanic_cleaned.csv', index=False)
```
